```python
class State(enum.IntEnum):
    SUSCEPTIBLE = 0
    VACCINATED = 1
    INFECTED_S = 2
    INFECTED_V = 3
    RECOVERED_S = 4
    RECOVERED_V = 5
```

```python
class MyAgent(Agent):
    """ An agent in an epidemic model."""
    def __init__(self, SV,unique_id, model):
        super().__init__(unique_id, model)
        self.SV = SV
        if self.SV == "S":
            self.state = State.SUSCEPTIBLE
        elif self.SV == "V":
            self.state = State.VACCINATED 
#         self.infection_time = 0
    def move(self):
        """Move the agent"""

        possible_steps = self.model.grid.get_neighborhood(
            self.pos,
            moore=True,
            include_center=False)
        new_position = self.random.choice(possible_steps)
        self.model.grid.move_agent(self, new_position)
    
    
    def status(self):
        """Check infection status"""
        
        if self.state == State.INFECTED_S: 
            recovery_rate_Is = self.model.gamma
            
            recovered = np.random.choice([0,1], p=[recovery_rate_Is,1-recovery_rate_Is])
            if recovered == 0:
                self.state = State.RECOVERED_S

                
        if self.state == State.INFECTED_V: 
            recovery_rate_Iv = self.model.gamma * self.model.delta
            
            recovered = np.random.choice([0,1], p=[recovery_rate_Iv,1-recovery_rate_Iv])
            if recovered == 0:
                self.state = State.RECOVERED_V

            

    def contact(self):
        """Find close contacts and infect"""
        
        cellmates = self.model.grid.get_cell_list_contents([self.pos])       
        if len(cellmates) > 1:

            for other in cellmates:
                
                    #If randomly not conntected with other agent
                if self.random.random() > self.model.beta:
                    continue
                    
                    #If Contatct Wtih Other agent
                if self.state is State.INFECTED_S and other.state is State.SUSCEPTIBLE:
                    rate = self.model.beta
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_S
   
                        
                if self.state is State.INFECTED_S and other.state is State.VACCINATED:   
                    rate = self.model.beta * (1-self.model.eta)-self.model.beta * (1-self.model.eta)*self.model.effectiveness
                    
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_V

    
                if self.state is State.INFECTED_V and other.state is State.SUSCEPTIBLE:
                    rate = self.model.beta
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_S

                    
                if self.state is State.INFECTED_V and other.state is State.VACCINATED:
                    rate = self.model.beta * (1-self.model.eta)-self.model.beta * (1-self.model.eta)*self.model.effectiveness
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_V
        
                    
    def step(self):
        self.status()
        self.move()
        self.contact()
```

```python
class SIRVModel(Model):
    """A model for infection spread."""
    
    def __init__(self, N=10,num_vaccination = 5,width=10, height=10,
                beta = 0.833,gamma = 1/3, delta = 3,eta = 0.3, effectiveness = 0.5):
        
        self.num_agents = N
        self.num_vaccination = num_vaccination
        
        
        #parameters
        self.beta = beta
        self.gamma = gamma
        self.delta = delta
        self.eta = eta
        self.effectiveness = effectiveness
        

        
        
        self.grid = MultiGrid(width, height, True)
        self.schedule = RandomActivation(self)



        self.running = True
        
        # Agent Creation 
        for i in range(self.num_agents):
            if i < self.num_vaccination:
                a = MyAgent("V",i, self)
            else:
                a = MyAgent("S", i, self)
            self.schedule.add(a)
            
            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))
            

            # If want to make some agent initially infected but here we take initially 
            if a.state == State.VACCINATED:
                infected = np.random.choice([0,1], p=[0.01,0.99])
                
                if infected == 0:
                    a.state = State.INFECTED_V

                
            elif a.state == State.SUSCEPTIBLE:
                infected = np.random.choice([0,1], p=[0.03,0.97])
                if infected == 0:
                    a.state = State.INFECTED_S

                
        # Collecting States of agent
        self.datacollector = DataCollector(
            #model_reporters={"Gini": compute_gini}, 
            agent_reporters={"State": "state"})
            
        
 
    
    def step(self):
        self.datacollector.collect(self)
        self.schedule.step()
```

```python
class IBRA_Agent(Agent):
    """ An agent in an epidemic model."""
    def __init__(self,st,unique_id, model):
        super().__init__(unique_id, model) 
        self.st = st       
        if self.st == 0:
            self.state = State.SUSCEPTIBLE
        elif self.st == 1:
            self.state = State.VACCINATED           
        elif self.st == 2:
            self.state = State.INFECTED_S            
        elif self.st == 3:
            self.state = State.INFECTED_V
            
        elif self.st == 4:
            self.state = State.RECOVERED_S
            
        elif self.st == 5:
            self.state = State.RECOVERED_V
            
    def move(self):
        """Move the agent"""

        possible_steps = self.model.grid.get_neighborhood(
            self.pos,
            moore=True,
            include_center=False)
        new_position = self.random.choice(possible_steps)
        self.model.grid.move_agent(self, new_position)
        
        
            
    def status(self):
        pass

        

    def contact(self):
        """Find close contacts and infect"""
        
        cellmates = self.model.grid.get_cell_list_contents([self.pos])       
        if len(cellmates) > 1:

            for other in cellmates:
                
                    #If randomly not conntected with other agent
                if self.random.random() > 0.833:
                    continue
                    
                    #If Contatct Wtih Other agent
                    # HV <- SFR
                if self.state is State.VACCINATED and other.state is State.SUSCEPTIBLE:
                    rate = 1/(1+ np.exp(-self.model.cr/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED
   
                     #HV <- FFR   
                if (self.state is State.VACCINATED and other.state is State.INFECTED_S) or (self.state is State.VACCINATED and other.state is  State.RECOVERED_S):
                    rate = 1/(1+ np.exp((1-self.model.cr)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED

                    # IV <- SFR
                if (self.state is State.INFECTED_V and other.state is State.SUSCEPTIBLE) or (self.state is State.RECOVERED_V and other.state is State.SUSCEPTIBLE):
                    rate = 1/(1+ np.exp((-1-self.model.cr)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED

                    # IV <- FFR
                tf = (self.state is State.INFECTED_V and other.state is State.INFECTED_S) or \
                     (self.state is State.INFECTED_V and other.state is State.RECOVERED_S) or \
                     (self.state is State.RECOVERED_V and other.state is State.INFECTED_S) or \
                     (self.state is State.RECOVERED_V and other.state is State.RECOVERED_S)
                if tf:
                    
                    rate = 1/(1+ np.exp((-self.model.cr)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED
                      # SFR <- HV  
                if self.state is State.SUSCEPTIBLE and other.state is State.VACCINATED:
                    rate = 1/(1+ np.exp((self.model.cr)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED
                   # SFR <- IV     
                if (self.state is State.SUSCEPTIBLE and other.state is State.INFECTED_V) or (self.state is State.SUSCEPTIBLE and other.state is State.RECOVERED_V):
                    rate = 1/(1+ np.exp((self.model.cr+1)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED
                   # FFR <- HV     
                if (self.state is State.INFECTED_S and other.state is State.VACCINATED) or (self.state is State.INFECTED_S and other.state is State.RECOVERED_S):
                    rate = 1/(1+ np.exp((self.model.cr-1)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED
                        
                tf = (self.state is State.INFECTED_S and other.state is State.INFECTED_V) or \
                     (self.state is State.INFECTED_S and other.state is State.RECOVERED_V) or \
                     (self.state is State.RECOVERED_S and other.state is State.INFECTED_V) or \
                     (self.state is State.RECOVERED_S and other.state is State.RECOVERED_V)         
                if tf:
                    rate = 1/(1+ np.exp((self.model.cr)/self.model.kappa))
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.VACCINATED
            
            
            
    def step(self):
        self.status()
        self.move()
        self.contact()
```

```python
class IB_RA(Model):
    
    def __init__(self,data,width=10, height=10,cr = 0, kappa = 0.1):
        
        
#       Agents
        self.data = data
        self.num_agents = len(self.data)
        
        
#        Parameters
        self.cr = cr
        self.kappa = kappa
        
    
        self.grid = MultiGrid(width, height, True)
        self.schedule = RandomActivation(self)
        self.running = True
        
        # Agent Creation 
        for i in range(self.num_agents):
            st = data.item(i)
            a = IBRA_Agent(st,i, self)
            self.schedule.add(a)
            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))

                
        self.datacollector = DataCollector(
            agent_reporters={"State": "state"})
        
    def step(self):
        self.datacollector.collect(self)
        self.schedule.step()
```

```python
class SecondMyAgent(Agent):
    """ An agent in an epidemic model."""
    def __init__(self, st,unique_id, model):
        super().__init__(unique_id, model)
        self.st = st
        
        if self.st == 0:
            self.state = State.SUSCEPTIBLE
        elif self.st == 1:
            self.state = State.VACCINATED      
        elif self.st == 2:
            self.state = State.INFECTED_S  
        elif self.st == 3:
            self.state = State.INFECTED_V
            
        elif self.st == 4:
            self.state = State.RECOVERED_S
            
        elif self.st == 5:
            self.state = State.RECOVERED_V

    
    def move(self):
        """Move the agent"""

        possible_steps = self.model.grid.get_neighborhood(
            self.pos,
            moore=True,
            include_center=False)
        new_position = self.random.choice(possible_steps)
        self.model.grid.move_agent(self, new_position)
    
    
    def status(self):
        """Check infection status"""
        
        if self.state == State.INFECTED_S: 
            recovery_rate_Is = self.model.gamma
            
            recovered = np.random.choice([0,1], p=[recovery_rate_Is,1-recovery_rate_Is])
            if recovered == 0:
                self.state = State.RECOVERED_S

                
        if self.state == State.INFECTED_V: 
            recovery_rate_Iv = self.model.gamma * self.model.delta
            
            recovered = np.random.choice([0,1], p=[recovery_rate_Iv,1-recovery_rate_Iv])
            if recovered == 0:
                self.state = State.RECOVERED_V

            

    def contact(self):
        """Find close contacts and infect"""
        
        cellmates = self.model.grid.get_cell_list_contents([self.pos])       
        if len(cellmates) > 1:

            for other in cellmates:
                
                    #If randomly not conntected with other agent
                if self.random.random() > self.model.beta:
                    continue
                    
                    #If Contatct Wtih Other agent
                if self.state is State.INFECTED_S and other.state is State.SUSCEPTIBLE:
                    rate = self.model.beta
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_S
   
                        
                if self.state is State.INFECTED_S and other.state is State.VACCINATED:   
                    rate = self.model.beta * (1-self.model.eta)*self.model.effectiveness
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_V

    
                if self.state is State.INFECTED_V and other.state is State.SUSCEPTIBLE:
                    rate = self.model.beta
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_S

                    
                if self.state is State.INFECTED_V and other.state is State.VACCINATED:
                    rate = self.model.beta * (1-self.model.eta)*self.model.effectiveness
                    decision = np.random.choice([0,1], p=[rate,1-rate])
                    
                    if decision == 0:
                        other.state = State.INFECTED_V
        
                    
    def step(self):
        self.status()
        self.move()
        self.contact()
```

```python
class SecondSIRVModel(Model):
    """A model for infection spread."""
    
    def __init__(self,data,width=10, height=10,
                beta = 0.833,gamma = 1/3, delta = 3,eta = 0.3, effectiveness = 0.5):
        
        # Agents 
        
        self.data = data
        self.num_agents = len(self.data)
        
        
        #parameters
        self.beta = beta
        self.gamma = gamma
        self.delta = delta
        self.eta = eta
        self.effectiveness = effectiveness
        

        
        
        self.grid = MultiGrid(width, height, True)
        self.schedule = RandomActivation(self)



        self.running = True
        
        # Agent Creation 
        for i in range(self.num_agents):
            st = data.item(i)
            a = SecondMyAgent(st, i, self)   
            
            self.schedule.add(a)
            
            # Add the agent to a random grid cell
            x = self.random.randrange(self.grid.width)
            y = self.random.randrange(self.grid.height)
            self.grid.place_agent(a, (x, y))
            
            
            # If want to make some agent initially infected but here we take initially 
            if a.state == State.VACCINATED or a.state == State.RECOVERED_V:
                infected = np.random.choice([0,1], p=[0.01,0.99])
                
                if infected == 0:
                    a.state = State.INFECTED_V

                
            elif a.state == State.SUSCEPTIBLE or a.state == State.RECOVERED_S:
                infected = np.random.choice([0,1], p=[0.03,0.97])
                if infected == 0:
                    a.state = State.INFECTED_S
                    
            
        self.datacollector = DataCollector(
            agent_reporters={"State": "state"})
            
        
 
    
    def step(self):
        self.datacollector.collect(self)
        self.schedule.step()
```

```python
def get_column_data(model):
    #pivot the model dataframe to get states count at each step
    agent_state = model.datacollector.get_agent_vars_dataframe()
    X = pd.pivot_table(agent_state.reset_index(),index='Step',columns='State',aggfunc=np.size,fill_value=0)    
    labels = ['Susceptible','Vaccinated','Infected_S','Infected_V','Recovered_S', 'Recovered_V']
    X.columns = labels[:len(X.columns)]
    return X
```

```python
def prep(model):
    X = get_column_data(model)
    X["Vaccinated"] = X["Vaccinated"] + X["Infected_V"] 
    X["Infected"] = X['Infected_S'] + X['Infected_V']
    X["Recoverd"] = X["Recovered_S"] + X["Recovered_V"]
    X.drop(columns=['Infected_S','Infected_V','Recovered_V','Recovered_S'], axis=1,inplace=True)
    X.to_csv("MergeColumn.csv")
    return X

X = prep(model)
f,ax= plt.subplots(1,1,figsize=(10,6))
X.plot(ax=ax,lw=3,alpha=0.8,xlim=[0,50],ylim=[0,16000])
plt.plot()
plt.savefig('SIRV0.jpg',dpi=500)
X.head()
```

```python
def get_data(data):
    d = []
    d.append(np.sum(data == 0))
    d.append(np.sum(data == 1))
    d.append(np.sum(data == 2))
    d.append(np.sum(data == 3))
    d.append(np.sum(data == 4))
    d.append(np.sum(data == 5))
    return d
```

```python

data = []
data.append(get_data(data1))

crs = np.arange(0,1,0.04)
effs = np.arange(0,1, 0.04)

# for relative cost
it = 0;
for cr in crs :
    for eff in effs:
        # Second Stage
        ibra = IB_RA(data = data1,width=110, height=110,cr = cr )
        for i in range(2):
            ibra.step()

        data2 = ibra.datacollector.get_agent_vars_dataframe().tail(30000).to_numpy() 
        
        # Again Epidemic
        model2 = SecondSIRVModel(data = data2, width= 110,height=110,eta = 0, effectiveness = eff)
        for i in range(step1):
            model2.step()
        data1 = model2.datacollector.get_agent_vars_dataframe().tail(30000).to_numpy()
        data.append(get_data(data1))
        print("iteration: ", it)
        it = it + 1
        
```
